```python
# DLL
class Node:
    def __init__(self, data):
        self.data = data
        self.f_link = None
        self.b_link = None

class DLL:
    def __init__(self):
        self.head = None
        self.index = 0


    def insert(self, location, data):   #O(N)
        # data 미입력 오류
        if data is None:
            raise ValueError("Cannot insert None value.")

        # 범위 오류 계산
        if location < 0 or location > self.index:
            raise IndexError("Out of the index.")

        new_node = Node(data)

        # case1: new data
        if self.head is None:
            self.head = new_node

        # case2: head 자리에 삽입
        elif location == 0:
            new_node.f_link = self.head
            self.head.b_link = new_node
            self.head = new_node

        # case3: tale 자리에 삽입
        elif location == self.index:
            current = self.head

            for i in range(location - 1):
                current = current.f_link
            new_node.b_link = current
            current.f_link = new_node

        # case4: Normal case
        else:
            current = self.head
            for i in range(location):
                current = current.f_link

            back_node = current.b_link
            new_node.b_link = back_node
            new_node.f_link = current
            current.b_link = new_node
            if back_node:
                back_node.f_link = new_node


        self.index += 1


    def delete(self, location):     #O(N)
        # 범위 오류 계산
        if location < 0 or location >= self.index:
            raise IndexError("Out of the index.")

        # head.link 데이터를 제거하는지 확인
        if location == 0:
            # head.f_link 가 None 인지 확인
            if self.head.f_link != None:
                self.head.f_link.b_link = None
                self.head = self.head.f_link
            else:
                return
        else:
            current = self.head
            for i in range(location - 1):
                current = current.f_link
            current.f_link = current.f_link.f_link
            if current.f_link is not None:
                current.f_link.b_link = current

        self.index -= 1


    def search(self, value):    #O(N)
        # value 값을 안 줬을 때
        if value is None:
            raise ValueError("Cannot insert None value.")

        current = self.head
        location = 0

        while current:
            if current.data != value:
                current = current.f_link
                location += 1
            else:
                break

        # value 값이 없을 때
        if location == self.index:
            return -1
        else:
            return location


    def reverse(self):      #O(N)
        # node 가 1개 이하 일 때
        if self.index <= 1:
            return

        back_node = None
        current = self.head

        while current:
            next_node = current.f_link
            current.f_link = back_node
            current.b_link = next_node
            back_node = current
            current = next_node
        self.head = back_node


    def swap(self, lo1, lo2):       #O(N)
        # 두 location 이 같음
        if lo1 == lo2:
            return
        # location 이 index 의 범위를 벗어남
        elif lo1 < 0 or lo2 < 0 or lo1 >= self.index or lo2 >= self.index:
            raise IndexError("Out of the index.")

        if lo1 > lo2:
            lo1, lo2 = lo2, lo1

        current1 = self.head
        for i in range(lo1):
            current1 = current1.f_link
        current2 = self.head
        for j in range(lo2):
            current2 = current2.f_link

        current1.data, current2.data = current2.data, current1.data


    def max(self):      #O(N)
        # 비어있는 리스트일 경우
        if self.index == 0:
            raise IndexError("Out of the index.")

        current = self.head
        max_value = current.data

        for i in range(self.index - 1):
            current = current.f_link
            if max_value < current.data:
                max_value = current.data

        return max_value


    def min(self):      #O(N)
        # 비어있는 리스트일 경우
        if self.index == 0:
            raise IndexError("Out of the index.")

        current = self.head
        min_value = current.data

        for i in range(self.index - 1):
            current = current.f_link
            if min_value > current.data:
                min_value = current.data

        return min_value


    def print_list(self):
        current = self.head
        result = []
        while current != None:
            result.append(current.data)
            current = current.f_link
        return result

# linkedlist 생성
ll = DLL()

# insert
ll.insert(0, 1)
ll.insert(0, 2)
ll.insert(1, 3)
ll.insert(3, 4)
ll.insert(4, 5)
ll.insert(5, 6)
ll.insert(6, 7)
ll.insert(7, 8)
ll.insert(8, 9)
ll.insert(9, 10)
print(ll.print_list())
print()

# delete
ll.delete(0)
ll.delete(3)
print(ll.print_list())
print()

# search
print(ll.search(3))
print(ll.search(100))
print()

# reverse
ll.reverse()
print(ll.print_list())
print()

# swap
ll.swap(0,7)
print(ll.print_list())
print()

#max
print(ll.max())
print()

#min
print(ll.min())
print()
```
